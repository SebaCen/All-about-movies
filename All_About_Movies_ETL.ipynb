# Transformaci√≥n

## Se acondiciona movies

### Cantidad de filas y columnas del dataframe

```
movies_original.shape
```

### Tipos de datos de cada columna

```
movies_original.dtypes
```

### Se realiza una copia para poder realizar el ETL y tener los datos originales para comparar

```
movies = movies_original.copy()
```

### Se convierte el Id a tipo de dato entero

```
def convertir_id_a_numerico(fila):

    try:
        return int(fila)
    except ValueError:
        return 0

movies['id'] = movies['id'].apply(convertir_id_a_numerico)
```

### Se eliminan Id duplicados

```
movies.drop_duplicates(subset='id', inplace=True)
```

### Se eliminan filas sin Id

```
movies.drop(movies[movies['id']==0].index, inplace=True)
```

### Se eliminan columnas que no aportan informacion relacionada al proyecto

```
movies.drop(['adult', 'video','imdb_id','original_title','poster_path','homepage'], axis=1, inplace=True)
```

### Se verifica el tama√±o del dataframe resultante

```
movies.shape
```

### Vista preliminar del dataframe resultante

```
movies
```

## Se acondiciona credits

### Se eliminan duplicados

```
credits.drop_duplicates(subset='id', inplace=True)
```

### Se verifica que no queden duplicados

```
credits['id'].value_counts()
```

### Se une el dataframe de "credits" al dataframe "movies" por el id en comun

```
movies = pd.merge(movies, credits, on='id', how='outer', indicator=True)
```

### Se eliminan nulos en la columna de protagonistas

```
movies.dropna(subset=['cast'], inplace=True)
```

### Se elimina columna de verificaci√≥n del merge

```
movies.drop('_merge', axis=1, inplace=True)
```

### Se convierte la columna presupuesto a tipo de dato flotante

```
movies['budget'] = pd.to_numeric(movies['budget'], errors='coerce').astype(float)
```

### Se convierte la columna recaudaci√≥n a tipo de dato flotante

```
movies['revenue'] = pd.to_numeric(movies['revenue'], errors='coerce').astype(float)
```

### Se reemplazan valores nulos por ceros en las columnas presupuesto y recaudaci√≥n

```
movies_revenue = movies['revenue']
movies_revenue.fillna(0, inplace=True)
movies_budget = movies['budget']
movies_budget.fillna(0, inplace=True)
```

### Se crea la columna retorno dividiendo la recaudaci√≥n por el presupuesto de cada pelicula

```
def crear_return(fila):
    if fila['budget']==0:
        return 0 
    else:
        return fila['revenue']/fila['budget']

movies['return'] = movies.apply(crear_return, axis=1)
```

### Se verifica si hay nulos en la columna fecha de lanzamiento

```
movies['release_date'].isnull().sum()
```

### Se eliminan esos nulos

```
movies.dropna(subset=['release_date'], inplace=True)
```

### Se convierte la fecha de lanzamiento a tipo de dato datetime

```
movies['release_date'] =pd.to_datetime(movies['release_date'], errors='coerce')
```

### Se crea la columna a√±o de lanzamiento

```
movies['release_year']=movies['release_date'].dt.year
```

### Se desanida la columna de pertenencia a una colecci√≥n

```
movies['belongs_to_collection'] = movies['belongs_to_collection'].apply(lambda fila:0 if type(fila)==float else fila)
movies['belongs_to_collection'] = movies['belongs_to_collection'].apply(lambda fila: fila[fila.find("'name'"):fila.find(", 'poster_path'")] if isinstance(fila, str) and 'name' in fila and 'poster_path' in fila else fila)
movies['belongs_to_collection'] = movies['belongs_to_collection'].apply(lambda fila: fila.replace("name': '","")  if isinstance(fila, str) else fila)
```

### Se desanida la columna generos

```
def desanidar_generos(fila):
    try:
        list = ast.literal_eval(fila)
        fila_aux = [fila_aux['name'] for fila_aux in list]
        return fila_aux
    except (SyntaxError, ValueError):
        return []
    
movies['genres']=movies['genres'].apply(desanidar_generos)
```

### Se verifica el tipo de dato de la columna generos

```
type(movies.loc[0,'genres'])
```

### Se desanida la columna companias de producci√≥n

```
def desanidar_companias(fila):
    try:
        list = ast.literal_eval(fila)
        fila_aux = [fila_aux['name'] for fila_aux in list]
        return fila_aux
    except (SyntaxError, ValueError):
        return []
    
movies['production_companies']=movies['production_companies'].apply(desanidar_companias)
```

### Se desanida la columna paises de producci√≥n

```
def desanidar_paises(fila):
    try:
        list = ast.literal_eval(fila)
        fila_aux = [fila_aux['name'] for fila_aux in list]
        return fila_aux
    except (SyntaxError, ValueError):
        return []
    
movies['production_countries']=movies['production_countries'].apply(desanidar_paises)
```

### Se desanida la columna de lenguajes

```
def desanidar_lenguajes(fila):
    try:
        list = ast.literal_eval(fila)
        fila_aux = [fila_aux['name'] for fila_aux in list]
        return fila_aux
    except (SyntaxError, ValueError):
        return []
    
movies['spoken_languages']=movies['spoken_languages'].apply(desanidar_lenguajes)
```

### Se desanida la columna protagonistas

```
def desanidar_personajes(fila):
    try:
        list = ast.literal_eval(fila)
        fila_aux = [fila_aux['character'] for fila_aux in list]
        fila_aux_2 = [fila_aux_2['name'] for fila_aux_2 in list]
        return [fila_aux,fila_aux_2]
    except (SyntaxError, ValueError):
        return []
    
movies['cast']=movies['cast'].apply(desanidar_personajes)
```

### Se crea un diccionario con los actores/actrices

```
def crear_diccionario(fila):
    diccionario = {}
    for key, value in zip(fila[0], fila[1]):
        diccionario[key] = value
    return diccionario

# Aplicar la funci√≥n a la columna del DataFrame y actualizar la columna con el diccionario
movies['cast'] = movies['cast'].apply(crear_diccionario)
```

### Se desanida la columna de equipo de filmaci√≥n

```
def desanidar_equipo(fila):
    try:
        list = ast.literal_eval(fila)
        fila_aux = [fila_aux['job'] for fila_aux in list]
        fila_aux_2 = [fila_aux_2['name'] for fila_aux_2 in list]
        return [fila_aux,fila_aux_2]
    except (SyntaxError, ValueError):
        return []
    
movies['crew']=movies['crew'].apply(desanidar_equipo)
```

### Se crea un diccionario con el equipo de filmaci√≥n

```
def crear_diccionario(fila):
    diccionario = {}
    for key, value in zip(fila[0], fila[1]):
        diccionario[key] = value
    return diccionario

movies['crew'] = movies['crew'].apply(crear_diccionario)
```

### Se crea la columna de directores  

```
movies['director'] = movies['crew'].apply(lambda x: x.get('Director', ''))
```

### Vista del dataframe con todas las columnas desanidadas y fin del proceso de limpieza

```
movies
```
